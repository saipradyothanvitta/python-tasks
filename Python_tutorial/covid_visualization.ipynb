```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

covid_df = pd.read_csv("/content/covid_data_updated.csv")
sns.set(style="whitegrid")
```

```python
covid_df['Date'] = pd.to_datetime(covid_df['Date'])
covid_cases_by_country = covid_df.groupby(['Date', 'Country'])['Confirmed'].sum().reset_index()

plt.figure(figsize=(12, 6))
sns.lineplot(data=covid_cases_by_country, x='Date', y='Confirmed', hue='Country', marker='o')
plt.title('Confirmed COVID-19 Cases Over Time by Country')
plt.xlabel("Date")
plt.ylabel("Confirmed Cases")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
covid_deaths_by_gender = covid_df.groupby('Gender')['Deaths'].sum().reset_index()

plt.figure(figsize=(6, 6))
sns.barplot(data=covid_deaths_by_gender, x='Gender', y='Deaths', palette='Set2')
plt.title('Total COVID-19 Deaths by Gender')
plt.xlabel("Gender")
plt.ylabel("Total Deaths")
plt.tight_layout()
plt.show()
```
