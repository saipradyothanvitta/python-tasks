```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

zomato_df = pd.read_csv("/content/zomato_sample.csv")
sns.set(style="whitegrid")
```

```python
avg_rating_by_city = zomato_df.groupby('City')['Aggregate Rating'].mean().sort_values(ascending=False).reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(data=avg_rating_by_city, x='City', y='Aggregate Rating', palette='coolwarm')
plt.title('Average Restaurant Rating by City')
plt.xlabel("City")
plt.ylabel("Average Rating")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=zomato_df, x='Average Cost for Two', y='Aggregate Rating', hue='City')
plt.title('Average Cost for Two vs. Aggregate Rating')
plt.xlabel("Average Cost for Two")
plt.ylabel("Aggregate Rating")
plt.tight_layout()
plt.show()
```
